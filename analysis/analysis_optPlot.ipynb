```python
#gaas = qtk.QMOut('gaas_ref/gaas_ref.out', program='abinit')
#alas = qtk.QMOut('gaas_rst_test/gaas_rst_test.out', program='abinit')
root = '/home/samio/Works/PhD/projects/01_AlGaAs/60_Eg/09_fcc333/'
os.chdir(root)
opt_log = qtk.Logger('optimization/algaas_333_01.db')
entries = [e for e in opt_log.list(has_data=True) if e.data > 0]
entries_sorted = [e for e in opt_log.list(has_data=True, order='descent') if e.data > 0]

mol_base = qtk.Molecule('xyz/gaas_2.xyz')
mol_test = mol_base.copy()
mol_base.extend([3,3,3], normalize=True)
mol_base.name = 'gaas_ref'
ccs = qtk.CCS(mol_base, 'ccs.yml')

def extract(i, entries_in = entries):
    
    def getOut(out_path):
        out = None
        if os.path.exists(out_path):
            out_file = glob.glob("%s/*.out" % out_path)[0]
            out = qtk.QMOut(out_file, program='abinit')
            return out
    
    out = None
    if i < len(entries_in):
        entry = entries_in[i]
        if 'processing' not in entry.comment:
            try:
                out_root = entry.comment.split("'")[1]
            except IndexError:
                out_root = entry.comment

            out_path = 'optimization/results/pred/%s' % out_root
            out_true = 'optimization/results/true/%s' % out_root
            
            #out_path = 'optimization/results/%s' % out_root
            #out_true = 'true_remote/%s' % out_root
            
    return getOut(out_path), getOut(out_true)
```

```python
def stepPlot(ax, ticksize=None, labelsize=None, legendsize=None, search_color='b', avg_color='g', ls = ':'):
    ax.plot(Eg_list, marker ='x', ls=ls, color=search_color, label='searched crystals', mec=search_color)
    ax.plot(max_list, color='k', lw=2)
    ax.step(max_x, max_y, where='post', 
             color='k', lw=2, marker='x', 
             markersize=10, mew=3, 
             label='best in population')
    ax.plot(avg_list, color=avg_color, lw=2, label='average of population')
    
    if labelsize:
        ax.set_xlabel('iterations (within the first GA search)', fontsize=labelsize)
        ax.set_ylabel(r'$E_g^{\mathrm{pred}}$ [eV]', fontsize=labelsize)
    else:
        ax.set_xlabel('iterations (within the first GA search)')
        ax.set_ylabel(r'$E_g^{\mathrm{pred}}$ [eV]')
    #ax.set_xticklabels(ax1.get_xticklabels(),fontsize=15)
    if ticksize:
        ax.set_xticklabels(range(0,len(entries),200),fontsize=ticksize)
        ax.set_yticklabels(np.arange(1, 1.7,0.1),fontsize=ticksize)
    else:
        ax.set_xticklabels(range(0,len(entries),200))
        ax.set_yticklabels(np.arange(1, 1.7,0.1))
        
    if legendsize:
        ax.legend(loc=4, fontsize=legendsize, numpoints=1)
    else:
        ax.legend(loc=4, numpoints=1)
    
    return ax

def bandPlot(ax, out, ticksize=None, labelsize=None, legendsize=None, zero_weight_color='k'):
    tick_pos = [0.031, 0.1043, 0.23]
    tick_txt = [
        r'$\mathrm{L}$', r'$\Gamma$', r'$\mathrm{X}$'
    ]
    out.plot_band_unfold([[0.5,0,0.25],[0,0,0],[0,0.5,0.5]], [4,4,4], 
                         ax=ax, zero_weight_color=zero_weight_color)
    ax.set_xticks(tick_pos)
    if ticksize:
        ax.set_xticklabels(tick_txt, fontsize=ticksize)
        ax.set_yticklabels(np.arange(-14,5,2), fontsize=ticksize)
    else:
        ax.set_xticklabels(tick_txt)
        ax.set_yticklabels(np.arange(-14,5,2))
    if labelsize:
        ax.set_xlabel(r'wavevector $\mathbf{k}$', fontsize=labelsize)
        ax.set_ylabel(r'$E(\mathbf{k})$ [eV]', fontsize=labelsize)
    else:
        ax.set_xlabel(r'wavevector $\mathbf{k}$')
        ax.set_ylabel(r'$E(\mathbf{k})$ [eV]')
    
    return ax
```

```python
import ast
import qctoolkit as qtk
import numpy as np
import matplotlib.pyplot as plt
plt.style.use('classic')

%matplotlib inline

gr_normalization = 'head'
#gr_normalization = None

mol_base = qtk.Molecule('xyz/gaas_2.xyz')
mol_base.extend([3,3,3], normalize=True)
mol_base.name = 'gaas_ref'
ccs = qtk.CCS(mol_base, 'ccs.yml')

ref_size = [6,6,6]
mol_base_ext = mol_base.copy()
mol_base_ext.extend(ref_size, normalize=True)

x, y_ref = mol_base_ext.gr('Ga', normalize=gr_normalization)
x, y_fcc = mol_base_ext.gr(normalize=gr_normalization)
plt.plot(x, y_ref)
plt.xlim(0,10)

# mol_test.extend([3,3,3], normalize=True)
# mol_test.extend([6,6,6], normalize=True)
# _x, _y = mol_test.gr('Ga')
# plt.figure()
# plt.plot(_x, _y)
# plt.xlim(0,10)

log01 = qtk.Logger('optimization/algaas_333_01.db')
log02 = qtk.Logger('optimization/algaas_333_02.db')
log03 = qtk.Logger('optimization/algaas_333_03.db')
log04 = qtk.Logger('optimization/algaas_333_04.db')

def gr(i, coords, elements=['Ga', 'Al']):
    mol = ccs.generate(**coords[i])
    mol.extend(ref_size, normalize=True)
    return mol.gr(*elements, normalize=gr_normalization)[1]

def get_gr_all(slice_start, slice_end, element_list, log):
    
    Eg, sorted_x = population_data(log)
    
    entries = log.list(order='descent')
    coords = []
    for e in entries:
        coords.append(ast.literal_eval(e.content))
    
    y_all = []
    for i in range(len(Eg))[slice_start:slice_end]:
        y = gr(i, coords, element_list)
        #y_all.append(y_ref - y)
        y_all.append(y)
    return np.stack(y_all)

def gr_mean_diff(gr1, gr2, plot=True):
    plt.figure()
    gr_diff = gr1.mean(0) - gr2.mean(0)
    if plot:
        plt.plot(x, gr_diff)
        plt.xlim([0,20])
    return gr_diff

def gr_fcc(y_homo):
    fig, [ax1, ax2, ax3] = plt.subplots(3,figsize=(8,6), sharex=True) 
    
    def plot_all():
        ax1.plot(x, y_homo / y_homo[y_homo > 0][0])
        ax1.set_xlim([0,20])
        ax2.plot(x, y_ref / y_ref[y_ref > 0][0])
        ax2.set_xlim([0,20])
        ax3.plot(x, y_homo / y_homo[y_homo > 0][0] - y_ref / y_ref[y_ref > 0][0])
        ax3.set_xlim([0,20])
        
    try:
        plot_all()
        
    except:
        y_homo = y_homo.mean(0)
        plot_all()
        
def get_normed(y):
    if len(y.shape) == 1:
        return np.asarray(y) / y[y>0][0]
    else:
        y = y.mean(0)
        return y / y[y>0][0]
```

```python
def population_data(log):
    entries = log.list(order='descent')
    coords = []
    for e in entries:
        if e.data > 0:
            coords.append(ast.literal_eval(e.content))
    Eg_list = [e.data for e in entries if e.data > 0]
    
    x_Al = []
    for coord in coords:
        mol = ccs.generate(**coord)
        x_Al.append(mol.stoichiometry(output='count')['Al'] / float(mol.N) * 2)
    
    return Eg_list, x_Al

def smear(x, y, sigma=0.05):
    x, y = np.asarray(x), np.asarray(y)
    ind = y != 0
    
    return (np.exp(-((x - x[ind][:, np.newaxis])/(2 * sigma)) ** 2) * y[ind][:, np.newaxis]).sum(0)

#gr_fcc = y_ref / max(y_ref)

plt.plot(x, get_normed(y_ref))
plt.xlim(0,20)
```

```python
def population_plot(log, new=False):
    
    Eg, sorted_x = population_data(log)

    #fig, [ax1, ax2] = plt.subplots(2,figsize=(16,12))
    fig, [ax1, ax2] = plt.subplots(2,figsize=(8,6))
    ax1.set_xlim([0,len(Eg)])
    ax1up = ax1.twinx()
    ax1up.set_xlim([0,len(Eg)])
    
    shift = 0
    range1, range2, range3 = np.arange(0, 50) + shift, np.arange(475, 525) + shift, np.arange(len(Eg)-50, len(Eg)) + shift
    
    ax1up.fill_between(range1, 2, facecolor='0.8', alpha=0.5)
    #ax1up.fill_between(range2, 2, facecolor='0.8', alpha=0.5)
    ax1up.fill_between(range3, 2, facecolor='0.8', alpha=0.5)

    ax1up.plot(Eg, lw=2)
    ax1.plot(sorted_x, color='0.3', ls=':', marker='x')
    
    ax1.yaxis.set_label_position('right')
    ax1.yaxis.tick_right()
    ax1.set_ylim(0.2, 0.8)
    
    ax1up.yaxis.set_label_position('left')
    ax1up.yaxis.tick_left()
    ax1up.set_ylim(0.9, 1.7)
    
    for ax in [ax1, ax1up]:
        ax.xaxis.set_label_position('top')
        ax.xaxis.tick_top()
        ax.set_xlabel('Crystal ID', fontsize=15)
        ax.xaxis.set_tick_params(labelsize=12)
        ax.yaxis.set_tick_params(labelsize=12)
    ax1.set_ylabel(r'$x$', fontsize=15)
    ax1up.set_ylabel(r'$E_g$ [eV]', fontsize=15)
    
    y_ref_normed = get_normed(y_ref)
    
    if new:
        gr_GaAl_top = get_gr_all(range1, ['Ga','Al'], log)
        gr_GaAl_bad = get_gr_all(range3, ['Ga','Al'], log)
    else:
        gr_GaAl_top = gr_GaAl_top_01
        gr_GaAl_bad = gr_GaAl_bad_01
        
    gr_top = get_normed(gr_GaAl_top)
    gr_bad = get_normed(gr_GaAl_bad)
    gr_GaAl_good_diff = gr_top - y_ref_normed
    gr_GaAl_bad_diff = gr_bad - y_ref_normed
    
    ind = np.asarray(y_ref_normed) != 0
    gr_ref = np.asarray(y_ref_normed)[ind]
    x_ref = np.asarray(x)[ind]
    
    max_good = max(abs(gr_GaAl_good_diff))
    max_bad = max(abs(gr_GaAl_bad_diff))

    ax2.plot(x_ref, gr_ref, color = 'k', marker = 'o', ls='')
    ax2.plot(x, y_ref_normed, color='k', lw=0.5)
    
    ax2.plot(x, smear(x, gr_GaAl_bad_diff), lw=2, color='r')
    ax2.plot(x_ref, gr_GaAl_bad_diff[ind], color='r', marker='^', ls='', markersize=8)
    
    ax2.plot(x, smear(x, gr_GaAl_good_diff), lw=2, color='g')
    ax2.plot(x_ref, gr_GaAl_good_diff[ind], color='g', marker='v', ls='', markersize=8)
    #ax2.plot(x, smear(x, gr_GaAl_good_diff) / max_good, lw=1, ls='--', color='g')
    #ax2.plot(x_ref, gr_GaAl_good_diff[ind] / max_good, color='g', marker='v', ls='', markersize=4)
    
    ax2.set_xlim(3, 15)
    ax2.set_ylim(-1.05, 1.05)
    ax2.set_ylabel(r'$g(r)$', fontsize=15)
    ax2.set_xlabel(r'$r$ [$\mathrm{\AA}$]', fontsize=15)
    ax2.xaxis.set_tick_params(labelsize=12)
    ax2.yaxis.set_tick_params(labelsize=12)
    
    def make_legend(ax):
        ax.plot([100,101], [100,101], color='b', lw=2, label = r'$E_g$')
        ax.plot([100,101], [100,101], color='0.3', marker='x', ls=':', label = r'$x$')
        ax.plot([100,101], [100,101], color='k', marker='o', lw=0.5, label = r'$g_{\mathrm{fcc}}(r)$')
        ax.plot([100,101], [100,101], color='g', marker='v', markersize=8, lw=2, 
                label = r'$\bar{g}_{\mathrm{b}}(r) - g_{\mathrm{fcc}}(r)$')
        ax.plot([100,101], [100,101], color='r', marker='^', markersize=8, lw=2, 
                label = r'$\bar{g}_{\mathrm{w}}(r) - g_{\mathrm{fcc}}(r)$')
        
        return ax
    
    ax = make_legend(ax1up)
    ax.legend(loc=(0.12, -0.35), numpoints=1, fontsize=15)
    return fig
    
fig = population_plot(log01)
```
